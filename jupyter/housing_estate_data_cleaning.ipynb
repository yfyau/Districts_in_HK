```python
import pandas as pd
import numpy as np
import json
from shapely.geometry import shape, Point
```

```python
df = pd.read_csv('./uncleaned dataset/housing_estates_hk.csv')
df
```

```python
new_df = df[['latitude', 'longitude', 'Estate', 'Population', 'Median Age', 'Median Monthly Domestic Household Income', 'Median Rent to Income Ratio', 'Average Domestic Household Size']]

new_df['District'] = 'null'

new_df
```

```python
with open('./uncleaned dataset/hk_districts.geojson') as f:
    js = json.load(f)

for index, row in new_df.iterrows():
    point = Point(row['longitude'], row['latitude'])
    
    for feature in js['features']:
        polygon = shape(feature['geometry'])
        if polygon.contains(point):
            new_df.at[index, 'District'] = feature['properties']['District']

new_df
```

```python
new_df['Median Rent to Income Ratio'].min()
```

```python
new_df.to_json('./cleaned dataset/housing_estates_hk.json', orient='index')
```
