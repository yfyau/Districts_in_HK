```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./Dataset/Primary and Secondary School.xlsx')
```

```python
df
```

```python
df['DISTRICT'].unique()
```

```python
df['ENGLISH CATEGORY'].unique()
```

```python
district_list = df['DISTRICT'].unique()
category_list = df['ENGLISH CATEGORY'].unique()
```

```python
new_df = pd.DataFrame({}, columns=np.concatenate((['District'], category_list)))

for district in district_list:
    new_row = [district]
    for category in category_list:
        new_row.append( len(df[ df['DISTRICT']==district ] [ df['ENGLISH CATEGORY']==category ].index) )
    new_df.loc[len(new_df)] = new_row

new_df
```

```python
new_df.to_csv('./Primary and Secondary School.csv',index=False)
```
